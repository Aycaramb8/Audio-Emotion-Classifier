# Process Self Collected Data

## Convert mp4 files to audio file (wav)

```python
import subprocess
from pathlib import Path
```

```python
mp4_dir = Path("./data/reallife/mp4/ENG")
wav_dir = Path("./data/reallife/wav/ENG")

wav_dir.mkdir(parents=True, exist_ok=True)

for mp4_file in mp4_dir.glob("*.mp4"):
    
    wav_file = wav_dir / (mp4_file.stem + ".wav")
    
    cmd = [
        "ffmpeg",
        "-i", str(mp4_file),
        "-ac", "1",          # mono
        "-ar", "48000",      # sample rate
        "-vn",               # ignore video
        "-y",
        str(wav_file)
    ]

    subprocess.run(cmd)
    print(f"Converted {mp4_file.name}")
```
